````python
import matplotlib.pyplot as plt

null = df.isnull().sum().sum()
not_null = df.size - null

plt.pie(
    [null, not_null],
    labels=["Null", "Not Null"],
    autopct="%1.1f%%"
)
plt.title("Null vs Not Null Values")
plt.show()
````
