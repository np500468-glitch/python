## Null vs Not Null Pie Chart

This code creates a pie chart showing the proportion of null and non-null values in the dataset. It helps visualize the overall data completeness.

````python
import matplotlib.pyplot as plt

null = df.isnull().sum().sum()
not_null = df.size - null

plt.pie(
    [null, not_null],
    labels=["Null", "Not Null"],
    autopct="%1.1f%%"
)
plt.title("Null vs Not Null Values")
plt.show()
````

## Box Plot of Price

This code creates a **box plot** for the **Price(Dollar)** column. It helps identify the distribution of prices, median, spread, and potential outliers in the dataset.

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.boxplot(df['Price(Dollar)'].dropna())

plt.title("Box Plot of Price")
plt.ylabel("Price (Dollar)")
plt.show()
````

+## Correlation Heatmap

This code creates a correlation heatmap using Matplotlib. It shows the relationship between numeric variables and helps identify positive and negative correlations in the dataset.

````python
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(6,5))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.show()
````

## Seaborn Correlation Heatmap

This code creates a correlation heatmap using the Seaborn library. It shows the correlation between numeric variables with values displayed inside each cell, helping to understand relationships between features.

````python



plt.figure(figsize=(6,5))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
````
